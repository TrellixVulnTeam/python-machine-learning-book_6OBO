# ハードクラスタリング

## K-means法(K 平均法)

```python
# データを作る
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=150, n_features=2, centers=3,  # centers=[１番目のクラスタの中心座標(x1, y1), 2つ目(x2, y2), ３つ目(x3, y3)]
                                  cluster_std=0.5,  # クラスタ内の標準偏差 リストにしてそれぞれのクラスで別々の標準偏差を指定できる.
                                  shuffle=True, random_state=0, center_box=(-10, 10)) # center_box引数はよくわからない
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(X[:, 0], X[:, 1], c='white', marker='o', s=50, edgecolors='black')
plt.grid()
plt.tight_layout()
#plt.savefig('./figures/spheres.png', dpi=300)
plt.show()
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, 
            init='random', 
            n_init=10,  # 十回初期値を変えてk-meansを行い，もっともSSEが低かったものを採用
            max_iter=300,
            tol=1e-04,
            random_state=0)
y_km = km.fit_predict(X)

plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s=50,
            c='lightgreen',
            marker='s',
            label='cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s=50,
            c='orange',
            marker='o',
            label='cluster 2')
plt.scatter(X[y_km == 2, 0],
            X[y_km == 2, 1],
            s=50,
            c='lightblue',
            marker='v',
            label='cluster 3')
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s=250,
            marker='*',
            c='red',
            label='centroids')
plt.legend()
plt.grid()
plt.tight_layout()
#plt.savefig('./figures/centroids.png', dpi=300)
plt.show()
```

## K-means++法

K-means法でセントロイドの初期値の決め方を改良したもの

```python
#  K-means++法

km = KMeans(n_clusters=3, 
            init='k-means++',  # 初期値の選び方をk-means++に指定
            n_init=10, 
            max_iter=300,
            tol=1e-04,
            random_state=0)
y_km = km.fit_predict(X) # predictで予測ラベルを返す
#km.fit(X)
#y_km = km.labels_ # 今の場合，未知データの変換ではないのでこちらでもOK
plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s=50,
            c='lightgreen',
            marker='s',
            label='cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s=50,
            c='orange',
            marker='o',
            label='cluster 2')
plt.scatter(X[y_km == 2, 0],
            X[y_km == 2, 1],
            s=50,
            c='lightblue',
            marker='v',
            label='cluster 3')
# 各クラスタのセントロイドをプロット
# 各クラスタには cluster_centers_アトリビュートでアクセスできる
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s=200,
            marker='*',
            c='red',
            label='centroids')
plt.legend()
plt.grid()
plt.tight_layout()
#plt.savefig('./figures/centroids.png', dpi=300)
plt.show()
          
```

## シルエット図

[Silhouette Coefficient](http://scikit-learn.org/stable/modules/clustering.html#silhouette-coefficient)

- a : The mean distance between a sample and all other points in the same class.
- b : The mean distance between a sample and all other points in the next nearest cluster.

$$
s = \frac{b - a}{\max (a, b)}
$$

```python
import numpy as np
from matplotlib import cm
from sklearn.metrics import silhouette_samples

km = KMeans(n_clusters=3, 
            init='k-means++', 
            n_init=10, 
            max_iter=300,
            tol=1e-04,
            random_state=0)
y_km = km.fit_predict(X) # ラベルを予測

cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(X, y_km, metric='euclidean') # データごとのシルエット係数を計算
y_ax_lower, y_ax_upper = 0, 0
yticks = []
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals) # 棒グラフのbottomの幅を求める
    color = cm.jet( i / n_clusters)
    # widthに配列を渡す場合,bottom 引数の配列の長さと width　引数の配列の長さがあってないといけない. 
    plt.barh(bottom=range(y_ax_lower, y_ax_upper), width=c_silhouette_vals, height=1.0,  
             edgecolor='none', color=color)

    yticks.append((y_ax_lower + y_ax_upper) / 2.)  # 棒グラフの底辺の半分にクラスタラベルを追加するための配列
    y_ax_lower += len(c_silhouette_vals) # 棒の幅を足す
    # y_ax_lower = y_ax_upper この上の行はこれでもOK
    
silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color="red", linestyle="--")  # シルエット係数の平均値をかく

plt.yticks(yticks, cluster_labels + 1) # 目盛りを編集
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')

plt.tight_layout()
# plt.savefig('./figures/silhouette.png', dpi=300)
plt.show()
```

# 階層的クラスタリング

## 凝集型クラスタリング(完全連結法)

```python

import pandas as pd
import numpy as np

np.random.seed(123)

variables = ['X', 'Y', 'Z']
labels = ['ID_0', 'ID_1', 'ID_2', 'ID_3', 'ID_4']

X = np.random.random_sample([5, 3])*10
df = pd.DataFrame(X, columns=variables, index=labels)
df
```

```python
"""
    scipy.spatial.distances.pdistはn次元空間でのデータ点ごとの距離を返す関数
    行方向にデータが積まれているとする．
    戻り値は距離行列の上三角行列のみである．これは距離行列が対象行列であり対角成分が
    ０のためである．距離行列にするにはscipy.spatial.distances.squareform()を使えば良い
    引数のmetricにはどのような距離を求めるかの指定もできる．
"""

from scipy.spatial.distance import pdist, squareform

row_dist = pd.DataFrame(squareform(pdist(df.values, metric='euclidean')), # ユークリッド距離を求める.
                                               columns=labels, index=labels)
row_dist
```

```python
from scipy.cluster.hierarchy import linkage
help(linkage)
```

[document](https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.linkage.html)

```python
# 正しくないコード 実行できるが警告が出る
from scipy.cluster.hierarchy import linkage

row_clusters = linkage(row_dist, method='complete', metric='euclidean')
pd.DataFrame(row_clusters,
             columns=['row label 1', 'row label 2',
                      'distance', 'no. of items in clust.'],
             index=['cluster %d' % (i + 1)
                    for i in range(row_clusters.shape[0])])
```

```python
# 2. correct approach: Condensed distance matrix

row_clusters = linkage(pdist(df.values, metric='euclidean'), method='complete') # linkage関数の入力に 上三角行列を入れる
pd.DataFrame(row_clusters,
             columns=['row label 1', 'row label 2',
                      'distance', 'no. of items in clust.'],
             index=['cluster %d' % (i + 1) 
                    for i in range(row_clusters.shape[0])])
```

```python
# 3. correct approach: Input sample matrix

row_clusters = linkage(df.values, method='complete', metric='euclidean') # linkage関数の入力に データ(行数=サンプル数)を入れる

pd.DataFrame(row_clusters,
             columns=['row label 1', 'row label 2',
                      'distance', 'no. of items in clust.'],
             index=['step %d' % (i + 1)
                    for i in range(row_clusters.shape[0])])
```

```python
#　デンドグラムを書く

from scipy.cluster.hierarchy import dendrogram

# make dendrogram black (part 1/2)
# from scipy.cluster.hierarchy import set_link_color_palette
# set_link_color_palette(['black'])

row_dendr = dendrogram(row_clusters, 
                       labels=labels, # 各データのlabelを指定
                       # make dendrogram black (part 2/2)
                       # color_threshold=np.inf
                       )
plt.tight_layout()
plt.ylabel('Euclidean distance')
#plt.savefig('./figures/dendrogram.png', dpi=300, 
#            bbox_inches='tight')
plt.show()
```

# DBSCAN を使ってクラスタリング

以下がアルゴリズム的にわかりやすい

https://ja.wikipedia.org/wiki/DBSCAN

http://scikit-learn.org/stable/modules/clustering.html#dbscan

```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
plt.scatter(X[:, 0], X[:, 1])
plt.tight_layout()
# plt.savefig('./figures/moons.png', dpi=300)
plt.show()
```

```python
# K-means と 完全連結法を用いてみる
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))

km = KMeans(n_clusters=2, random_state=0)
y_km = km.fit_predict(X)
ax1.scatter(X[y_km == 0, 0], X[y_km == 0, 1],
            c='lightblue', marker='o', s=40, label='cluster 1')
ax1.scatter(X[y_km == 1, 0], X[y_km == 1, 1],
            c='red', marker='s', s=40, label='cluster 2')
ax1.set_title('K-means clustering')

ac = AgglomerativeClustering(n_clusters=2,
                             affinity='euclidean',
                             linkage='complete')
y_ac = ac.fit_predict(X)
ax2.scatter(X[y_ac == 0, 0], X[y_ac == 0, 1], c='lightblue',
            marker='o', s=40, label='cluster 1')
ax2.scatter(X[y_ac == 1, 0], X[y_ac == 1, 1], c='red',
            marker='s', s=40, label='cluster 2')
ax2.set_title('Agglomerative clustering')

plt.legend()
plt.tight_layout()
#plt.savefig('./figures/kmeans_and_ac.png', dpi=300)
plt.show()
```

```python
# Density-based clustering:

from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.2, # 隣接点とみなす２点間の最大の距離
            min_samples=5, # ボーダー点と判断される最小個数
            metric='euclidean') #  距離の計算方法
y_db = db.fit_predict(X)
plt.scatter(X[y_db == 0, 0], X[y_db == 0, 1],
            c='lightblue', marker='o', s=40,
            label='cluster 1')
plt.scatter(X[y_db == 1, 0], X[y_db == 1, 1],
            c='red', marker='s', s=40,
            label='cluster 2')
plt.legend()
plt.tight_layout()
#plt.savefig('./figures/moons_dbscan.png', dpi=300)
plt.show()
```
