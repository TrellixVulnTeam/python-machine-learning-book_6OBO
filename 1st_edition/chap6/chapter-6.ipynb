```python
import pandas as pd
import urllib

try:
    df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases'
                     '/breast-cancer-wisconsin/wdbc.data', header=None)

except urllib.error.URLError:
    df = pd.read_csv('https://raw.githubusercontent.com/rasbt/'
                     'python-machine-learning-book/master/code/'
                     'datasets/wdbc/wdbc.data', header=None)
    
print('rows, columns:', df.shape)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
X = df.iloc[:, 2:].values
y = df.iloc[:, 1].values
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
le.classes_　# b: nagative, m:positive
```

```python
le.inverse_transform(y)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=None) 
# stratify 引数もあり,defaultでFalse , stratify=yとするとクラスラベルとしてこれを使用して、データが層別に分割されます。
print(np.unique(y_train, return_counts=True)[1]/ y_train.size)
print(np.unique(y_test, return_counts=True)[1]/ y_test.size)
print(y_train.size)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

estimators=[('scl', StandardScaler()), ('pca', PCA(n_components=2)), ('clf', LogisticRegression(random_state=1))]
pipe_lr = Pipeline(steps=estimators)
pipe_lr.fit(X_train, y_train)
print('Test Accuracy: {:.3f}'.format(pipe_lr.score(X_test, y_test)))
```

```python
from sklearn.model_selection import StratifiedKFold

kfold = StratifiedKFold(n_splits=10, random_state=1)

scores=[]
for  k, (idx_train, idx_test) in enumerate(kfold.split(X_train, y_train)):
    pipe_lr.fit(X_train[idx_train], y_train[idx_train])
    score=pipe_lr.score(X_train[idx_test], y_train[idx_test])
    scores.append(score)
    print('Fold: %s, Class dist.: %s, Acc: %.3f' % (k+1,
          np.bincount(y_train[idx_train]), score))
    
print('\nCV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(estimator=pipe_lr,
                         X=X_train,
                         y=y_train,
                         cv=10, # 層化の10-fold cross validation
                         n_jobs=1)
print('CV accuracy scores: %s' % scores)
print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```

```python
import matplotlib.pyplot as plt

from sklearn.model_selection import learning_curve

pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('clf', LogisticRegression(penalty='l2', random_state=0))])

"""
learning_curve()は学習曲線を返すための関数

引数
--------------------
estimator: 識別器または，パイプライン
train_sizes: リスト(サイズ：cv, 中身はそれぞれ学習データ数を表す．)
cv: int でStratified k-fold cross validation のk を指定できる.

戻り値
--------------------
train_size: shape=(train_sizes, ) 
学習データ数が格納されてある.パラメータで指定したtrain_sizeの重複部分を取り除いたもの.
train_scores, test_scores : shape=(train_sizes, cv)
行ごとに様々なサイズのトレーニングデータセットに対してのcv-fold交差検証の結果がはいっている．
"""
train_sizes, train_scores, test_scores =\
                learning_curve(estimator=pipe_lr,
                               X=X_train,
                               y=y_train,
                               train_sizes=np.linspace(0.1, 1.0, 10), # 0.1~0.99までデータ数の割合を変化させる
                               cv=10,
                               n_jobs=1)

train_mean = np.mean(train_scores, axis=1) # 行（学習データの割合）ごとの平均をとる
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

plt.plot(train_sizes, train_mean,
         color='blue', marker='o',
         markersize=5, label='training accuracy')

# y1 からy2 の間(平均+-標準偏差)の値を塗りつぶす
plt.fill_between(x=train_sizes,
                 y1=train_mean + train_std,
                 y2=train_mean - train_std,
                 alpha=0.15, color='blue')

plt.plot(train_sizes, test_mean,
         color='green', linestyle='--',
         marker='s', markersize=5,
         label='validation accuracy')

plt.fill_between(x=train_sizes,
                 y1=test_mean + test_std,
                 y2=test_mean - test_std,
                 alpha=0.15, color='green')

plt.grid()
plt.xlabel('Number of training samples')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.ylim([0.8, 1.0])
plt.tight_layout()
# plt.savefig('./figures/learning_curve.png', dpi=300)
plt.show()
```

```python
from sklearn.model_selection import validation_curve

param_range = [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]

"""validation_curve
モデルのパラメータを変化させて層化交差検証による正解率を算出する

-----------------------注意する引数
param_name: str
変化させるパラメータの名前
piplineを用いる時は'(変化させるパラメータの識別器の名前)__(変化させるパラメータの名前)'とする

param_range: array-like, shape (n_values,)
変化させるパラメータの値をいれたリスト
検証曲線のx軸となる

Return
-----------------------
learning_curveと違いtrain_scoresとtest_scoresしか返ってこない
"""
train_scores, test_scores = validation_curve(
                estimator=pipe_lr, 
                X=X_train, 
                y=y_train, 
                param_name='clf__C', 
                param_range=param_range,
                cv=10)

train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

plt.plot(param_range, train_mean, 
         color='blue', marker='o', 
         markersize=5, label='training accuracy')

plt.fill_between(param_range, train_mean + train_std,
                 train_mean - train_std, alpha=0.15,
                 color='blue')

plt.plot(param_range, test_mean, 
         color='green', linestyle='--', 
         marker='s', markersize=5, 
         label='validation accuracy')

plt.fill_between(param_range, 
                 test_mean + test_std,
                 test_mean - test_std, 
                 alpha=0.15, color='green')

plt.grid()
plt.xscale('log')
plt.legend(loc='lower right')
plt.xlabel('Parameter C')
plt.ylabel('Accuracy')
plt.ylim([0.8, 1.0])
plt.tight_layout()
# plt.savefig('./figures/validation_curve.png', dpi=300)
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

pipe_svc = Pipeline([('scl', StandardScaler()),
            ('clf', SVC(random_state=1))])

param_range = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]

param_grid = [{'clf__C': param_range, 
               'clf__kernel': ['linear']},
                 {'clf__C': param_range, 
                  'clf__gamma': param_range, 
                  'clf__kernel': ['rbf']}]

gs = GridSearchCV(estimator=pipe_svc, 
                  param_grid=param_grid, 
                  scoring='accuracy', # 性能評価にaccuracy_scoreを使うということ
                  cv=10,
                  n_jobs=-1, verbose=2)
gs = gs.fit(X_train, y_train)
print(gs.best_score_)
print(gs.best_params_)
```

```python
# 5x2 cross validation
gs = GridSearchCV(estimator=pipe_svc,
                  param_grid=param_grid,
                  scoring='accuracy',
                  cv=2)

# Note: Optionally, you could use cv=2 
# in the GridSearchCV above to produce
# the 5 x 2 nested CV that is shown in the figure.

scores = cross_val_score(gs, X_train, y_train, scoring='accuracy', cv=5)
print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```

```python
# 5x2 cross validation
from sklearn.tree import DecisionTreeClassifier

gs = GridSearchCV(estimator=DecisionTreeClassifier(random_state=0),
                  param_grid=[{'max_depth': [1, 2, 3, 4, 5, 6, 7, None]}],
                  scoring='accuracy',
                  cv=2)
scores = cross_val_score(gs, X_train, y_train, scoring='accuracy', cv=5)
print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```

ROC曲線について

```python
from sklearn.metrics import roc_curve, auc
from scipy import interp

pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('clf', LogisticRegression(penalty='l2', 
                                               random_state=0, 
                                               C=100.0))])
# 特徴を２個取ってくる
X_train2 = X_train[:, [4, 14]]

# 下でlen()関数を用いて長さをとり出したいのでリストにした
cv = list(StratifiedKFold(n_splits=3, 
                        random_state=1).split(X_train, y_train))

fig = plt.figure(figsize=(7, 5))

mean_tpr = 0.0
mean_fpr = np.linspace(0, 1, 100)
all_tpr = []

for i, (train, test) in enumerate(cv):
    probas = pipe_lr.fit(X_train2[train],
                         y_train[train]).predict_proba(X_train2[test])

    fpr, tpr, thresholds = roc_curve(y_true=y_train[test],
                                     y_score=probas[:, 1], # y_score引数にはdecision_functionかpredict_probaを指定. 詳しくは特徴エンジニアリングP285参照
                                     pos_label=1)
    # FPR(X軸)とTPR(Y軸)で線形補間
    """scipy.interp
    線形補間する関数
    
    注意すべきパラメータ
    -------------------------
    x: 補完したいデータ点が入った一次元リスト
    xp: 補完する元となるデータポインタが入ったリスト x座標
    fp: 補完する元となるデータポインタが入ったリスト y座標
    
    返り値
    -------------------------
    xに対応する補完されたデータ
    """
    mean_tpr += interp(x=mean_fpr, xp=fpr, fp=tpr) # 3-foldの平均を取るために合計を取る

    roc_auc = auc(fpr, tpr) # 曲線下面積を求める
    # ROC曲線をプロット
    plt.plot(fpr,
             tpr,
             lw=1,
             label='ROC fold %d (area = %0.2f)'
                   % (i+1, roc_auc))

plt.plot([0, 1],
         [0, 1],
         linestyle='--',
         color=(0.6, 0.6, 0.6),
         label='random guessing')

mean_tpr /= len(cv)
mean_tpr[0] = 0.0
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
plt.plot(mean_fpr, mean_tpr, 'k--',
         label='mean ROC (area = %0.2f)' % mean_auc, lw=2)
plt.plot([0, 0, 1],
         [0, 1, 1],
         lw=2,
         linestyle=':',
         color='black',
         label='perfect performance')

plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('Receiver Operator Characteristic')
plt.legend(loc="lower right")

plt.tight_layout()
# plt.savefig('./figures/roc.png', dpi=300)
plt.show()
```

高階関数について

```python
def logger(func):
    def inner(*args):
        print("引数", args)
        return func(*args)
    return inner
```

```python
@logger
def accumulate(a, b):
    return a+b
print(accumulate(1, 2))
```

```python
newfunc = logger(accumulate)
print(newfunc(1, 2))
```
